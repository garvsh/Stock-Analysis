```python
fb['LogReturn'] = np.log(fb['Close']).shift(-1) - np.log(fb['Close'])
```

```python
# Plot a histogram to show the distribution of log return of facebook's stock. 
# You can see it is very close to a normal distribution
from scipy.stats import norm
mu = fb['LogReturn'].mean()
sigma = fb['LogReturn'].std(ddof=1)

density = pd.DataFrame()
density['x'] = np.arange(fb['LogReturn'].min()-0.01, fb['LogReturn'].max()+0.01, 0.001)
density['pdf'] = norm.pdf(density['x'], mu, sigma)

fb['LogReturn'].hist(bins=50, figsize=(15, 8))
plt.plot(density['x'], density['pdf'], color='red')
plt.show()
```

```python
# probability that the stock price of facebook will drop over 5% in a day
prob_return1 = norm.cdf(-0.05, mu, sigma)
print('The Probability is ', prob_return1)
```

```python
# Now is your turn, calculate the probability that the stock price of microsoft will drop over 10% in a day
prob_return1 = norm.cdf(-0.1,mu,sigma)
print('The Probability is ', prob_return1)
```

```python
# drop over 40% in 220 days
mu220 = 220*mu
sigma220 = (220**0.5) * sigma
print('The probability of dropping over 40% in 220 days is ', norm.cdf(-0.4, mu220, sigma220))
```

# caluculate VaR

```python
VaR = norm.ppf(0.05, mu, sigma)
print('Single day value at risk ', VaR)
```

```python
# Quatile 
# 5% quantile
print('5% quantile ', norm.ppf(0.05, mu, sigma))
# 95% quantile
print('95% quantile ', norm.ppf(0.95, mu, sigma))
```

# Confidence level testing


```python
sample_size=fb['LogReturn'].shape[0]
sample_mean=fb['LogReturn'].mean()
sample_std=fb['LogReturn'].std(ddof=1)/sample_size**0.5
# 90% confidence level so alpha = 0.05

z_left=norm.ppf(0.05)
z_right=norm.ppf(0.95)
```

```python
interval_left = sample_mean + z_left*sample_std
interval_right = sample_mean + z_right*sample_std
```

```python

print("confidence level is ( {} , {} )".format(interval_left,interval_right))
```

# hypothesis testing

```python
# two tailed test
sample_mean = fb['LogReturn'].mean()
sample_std = fb['LogReturn'].std(ddof=1)
n = fb['LogReturn'].shape[0]
zhat=sample_mean/(sample_std/n**0.5)
print(zhat)
```

# decision criteria

```python
# confidence level
alpha = 0.05

zleft = norm.ppf(alpha/2, 0, 1)
zright = -zleft  # z-distribution is symmetric 
print(zleft, zright)
```

```python
print('At significant level of {}, shall we reject: {}'.format(alpha, zhat>zright or zhat<zleft))
```

```python
# lets try to guess what are the real values of intercept and slope
# we call our guess b0, b1...
# Try to assign the value of b0, b1 to get a straight line that can describe our data 
b0 = 0.1
b1 = 0.2
fb['guess'] = b0 + b1*fb['Close']
```

```python
fb['obserror']=fb['wealth'] - fb['guess']

plt.title('Sum of sqaured error is {}'.format((((fb['obserror'])**2)).sum()))
plt.scatter(fb['Close'], fb['wealth'], color='g', label='Observed')
plt.plot(fb['Close'], fb['guess'], color='red', label='GuessResponse')
plt.legend()
```

```python
fb.dropna(subset = ["wealth"], inplace=True)
ft = np.polyfit(fb.Close, fb.wealth, 1)
ft
```

```python

fb['best_resp']=ft[1] + ft[0]*fb['Close']
fb['error']=fb['wealth']-fb['best_resp']

plt.title('Sum of sqaured error is {}'.format((((fb['error'])**2)).sum()))
plt.scatter(fb['Close'],fb['wealth'],color='b',label='Observed')
plt.plot(fb['Close'],fb['best_resp'],color='r',label='Best Response')
```
